```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.watts_strogatz_graph(100, 10, 0.2)
nx.draw(G, with_labels=True)
```

```python
G = nx.barbell_graph(5, 8)
nx.draw(G, with_labels=True)
```

```python
path = nx.shortest_path(G, 1, 16)
print(path)
```

```python
path = nx.astar_path(G, 1, 16)
print(path)
```

```python
G = nx.karate_club_graph()
nx.draw(G, with_labels=True)
```

```python
paths = nx.shortest_simple_paths(G, 1, 16)
K = 5
for k, p in enumerate(paths):
    print(p)
    if k == K - 1:
      break
```

```python
paths = nx.all_pairs_shortest_path(G)
for k, p in enumerate(paths):
  print(k, p)
```

```python
paths = nx.single_source_shortest_path_length(G, 2)
for k, p in enumerate(paths):
  print(f"index:{k}, reachable node:{p}, paths length:{paths[p]}")
```

```python
mst = nx.minimum_spanning_tree(G)
nx.draw(mst, with_labels=True)
```
